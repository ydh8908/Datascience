```python
import pandas as pd
import requests

def price_data(firm_code):
    for num in range(1,500) :
        price_url = 'https://finance.naver.com/item/sise_day.nhn?code=' +firm_code +'&'+'page=' +str(num)
        price_page=requests.get(price_url)
        price_tables=pd.read_html(price_page.text)
        price_tables=price_tables[0]
        price_tables=price_tables.set_index(price_tables.columns[0])
        price_tables=price_tables[['종가']]
        price_tables.columns=[firm_code]
        filtered_price=price_tables.dropna()
        filtered_price=pd.DataFrame(filtered_price)
        if num==1:
            historic_price=filtered_price
        else:
            historic_price=pd.concat([ historic_price, filtered_price])

    return historic_price

path=r'/Users/donghyunyoo/Downloads/firm_code_list.xls'
code_data = pd.read_excel(path)
code_data= code_data[['종목코드']]

def make_code2(x):
    x=str(x)
    return '0' * (6-len(x))+x

firm_code_list=code_data['종목코드'].apply(make_code2)


for num, firm_code in enumerate(firm_code_list):
    print(num)
    individual_historic_price_data = price_data(firm_code)
    if num == 0:
        total_price = individual_historic_price_data
    else :
        total_price = pd.merge(total_price, individual_historic_price_data, how='outer', right_index=True, left_index=True)
    total_price.index = pd.to_datetime(total_price.index)

total_price


```
