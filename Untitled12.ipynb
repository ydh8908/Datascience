```python
import pandas as pd 
df = pd.read_csv('enrolment_2.csv')
df['room assignment']='Auditorium'
al = df.loc[df['status']=='allowed','course name'].value_counts()
auditorium = list((al[al>=80]).index)
largeroom = list(al[(al>=40) & (al<80)].index)
mediumroom = list(al[(al>=15) & (al<40)].index)
smallroom = list(al[(al>=5) & (al<15)].index)

for course in auditorium:
    df.loc[(df['course name']==course) & (df['status']=='allowed'), 'room assignment'] = 'auditorium'

for course in largeroom:
    df.loc[(df['course name']==course) & (df['status']=='allowed'), 'room assignment'] = 'largeroom'

for course in mediumroom:
    df.loc[(df['course name']==course) & (df['status']=='allowed'), 'room assignment'] = 'mediumroom'

for course in smallroom:
    df.loc[(df['course name']==course) & (df['status']=='allowed'), 'room assignment'] = 'smallroom'

df




    
# df.loc[auditorium, 'room assignment']='auditorium'
# df.loc[largeroom, 'room assignment']=='large room'
# df.loc[mediumroom, 'room assignment']=='medium room'
# df.loc[smallroom, 'room assignment']=='small room'
# df.to_excel('hello.xlsx')
```
