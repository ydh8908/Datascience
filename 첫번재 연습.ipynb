```python
def magic(file_path):
    import xlrd
    import operator

    wb=xlrd.open_workbook(file_path)

    per_sh=wb.sheet_by_name('PER')

    per_dict={}
    for i in range(1,per_sh.nrows):
        data=per_sh.row_values(i)
        name=data[0]
        per=data[1]
        if per>0:
            per_dict[name]=per

    sorted_per = sorted(per_dict.items(),key=operator.itemgetter(1))

    per_rank={}
    for num, firm in enumerate(sorted_per):
        per_rank[firm[0]]=num+1

    roa_sh = wb.sheet_by_name('ROA')

    roa_dict={}
    for i in range(1, roa_sh.nrows):
        data=roa_sh.row_values(i)
        name=data[0]
        roa=data[1]
        if roa!='':
            roa_dict[name]=roa

    sorted_roa=sorted(roa_dict.items(), key=operator.itemgetter(1), reverse=True)

    roa_rank={}
    for num, firm in enumerate(sorted_roa):
        roa_rank[firm[0]]=num+1

    total_rank ={}
    for name in roa_rank.keys():
        if name in per_rank.keys():
            total_rank[name]=per_rank[name]+roa_rank[name]

    sorted_total = sorted(total_rank.items(), key=operator.itemgetter(1))

    magic_rank={}
    for num, firm in enumerate(sorted_total):
        magic_rank[firm[0]]=num+1
    
    return magic_rank
```

```python
import pandas
file_path=('/Users/donghyunyoo/Downloads/hello.xlsx')
per_data=pd.read_excel(file_path, sheet_name='PER', index_col=0)
filtered_per = per_data[per_data['PER']>0]
sorted_per = filtered_per.sort_values('PER', ascending=True)
sorted_per['PER rank']=sorted_per['PER'].rank()

roa_data=pd.read_excel(file_path, sheet_name='ROA', index_col=0)
filtered_roa=roa_data.dropna()
filtered_roa.columns=['ROA']
sorted_roa = filtered_roa.sort_values(by='ROA', ascending= False)
sorted_roa['ROA rank']=sorted_roa['ROA'].rank(ascending = False)

total_df = pd.merge(sorted_per, sorted_roa, how='inner', left_index=True, right_index=True)
total_df['종합 랭크']=(total_df['PER rank'] + total_df['ROA rank']).rank()
total_df.sort_values(by='종합 랭크')
```

```python

#짧게 결과만 확인하는 용도
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(980707)

MaxLoans = 100 #최대 기초자산 수
SimRuns = 50 #포트폴리오별 시뮬레이션 횟수 ... 참고: 개별 채권 시뮬레이션 횟수는 (MaxLoans^2+MaxLoans)*SimRuns/2이므로 주의(느려짐)
cgArray = [1,2,3,4,5,6,7]
cgNice = [0.0911, 0.1589, 0.2098, 0.3478, 0.1730, 0.0192, 0.0002] #2018.10 ~ 2019.9 피플펀드 NICE 등급별 분포율(대출액기준)
pdNice = [0.0005, 0.0016, 0.0034, 0.0051, 0.0070, 0.0182, 0.0629] #2017년말 NICE 1~7등급 채무자의 2018년말 불량률 (NICE 공시)
irNice = [0.0786, 0.0989, 0.1094, 0.1166, 0.1454, 0.1515, 0.1549] #피플펀드 NICE 1~7등급별 가중평균금리(9월말 기준 최근 1년)
recoup = 0.2

Result = np.zeros((MaxLoans, SimRuns))

plt.figure(figsize=(10,5))
plt.xlim(0,MaxLoans); plt.xlabel('Number of Loans')
plt.ylim(-1,0.3); plt.ylabel('Rate of Return')
plt.xticks(np.arange(0, MaxLoans+1, MaxLoans/5))
plt.yticks(np.arange(-1, 0.31, 0.1))
plt.grid()

for NumLoans in range(1, MaxLoans + 1):
   LoanRet = np.zeros(NumLoans)
   for SimCnt in range(0, SimRuns):
       cgRan = np.random.choice(cgArray,NumLoans,p=cgNice)
       irRan = np.random.randn(NumLoans)
       pdRan = np.random.rand(NumLoans)
       for LoanCnt in range(0, NumLoans):
           cgLoan = int(cgRan[LoanCnt])
           if pdRan[LoanCnt] > pdNice[cgLoan-1]:
               LoanRet[LoanCnt] = irNice[cgLoan-1] + (irRan[LoanCnt] * 2) / 100
           else:
               LoanRet[LoanCnt] = (recoup + np.random.randn(1) / 100 * 2) - 1
       Result[NumLoans-1][SimCnt] = ret = np.mean(LoanRet)
       if ret > 0:
           plt.plot(NumLoans, ret, 'o', ms=5, c='red', alpha=0.02)
       else:
           plt.plot(NumLoans, ret, 'o', ms=5, c='blue', alpha=0.5)

plt.show()
```
